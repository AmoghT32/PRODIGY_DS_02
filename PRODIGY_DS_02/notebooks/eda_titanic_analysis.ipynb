# Titanic Dataset – Data Cleaning and Exploratory Data Analysis

This notebook performs data cleaning and exploratory data analysis on the Titanic dataset to identify patterns related to passenger survival.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
data_path = "../data/raw/titanic.csv"

df = pd.read_csv(data_path)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/titanic_cleaned.csv", index=False)
```

```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count")

plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival by Gender")

plt.show()
```

```python
sns.histplot(df['Age'], bins=30)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title("Survival by Passenger Class")

plt.savefig("../outputs/figures/survival_by_class.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)

plt.title("Age Distribution by Survival")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Pclass', y='Age', data=df)

plt.title("Age Distribution Across Passenger Classes")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Pclass', y='Age', data=df)

plt.title("Age Distribution Across Passenger Classes")

plt.savefig("../outputs/figures/age_boxplot.png", dpi=300)

plt.show()
```

## Key Insights

- Female passengers had a significantly higher survival rate compared to males.
- Passengers in first class had better survival chances than those in lower classes.
- Younger passengers had slightly higher survival probability.
- Passenger class and gender were strong factors influencing survival outcomes.
